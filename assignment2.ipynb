```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='â„–':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')
# df.head()
df
```

### Question 7
Which county has had the largest absolute change in population within the period 2010-2015? (Hint: population values are stored in columns POPESTIMATE2010 through POPESTIMATE2015, you need to consider all six columns.)

e.g. If County Population in the 5 year period is 100, 120, 80, 105, 100, 130, then its largest change in the period would be |130-80| = 50.

*This function should return a single string value.*

```python
counties_only = census_df[census_df['SUMLEV'] == 50]
required_columns = ['STNAME',
                    'CTYNAME',
                    'POPESTIMATE2010',
                    'POPESTIMATE2011',
                    'POPESTIMATE2012',
                    'POPESTIMATE2013',
                    'POPESTIMATE2014',
                    'POPESTIMATE2015']
counties_only = counties_only[required_columns]
counties_only.head()
```

```python
params = ['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']
counties_only['abs_change_pop'] = (counties_only[params].max(axis=1)) - (counties_only[params].min(axis=1))
counties_only
```

```python
ans = counties_only[counties_only['abs_change_pop'] == max(counties_only['abs_change_pop'])]
ans
```

```python
ans.iloc[0]['CTYNAME']
```

```python
counties_only = census_df[census_df['SUMLEV'] == 50]
params = ['POPESTIMATE2010',
          'POPESTIMATE2011',
          'POPESTIMATE2012',
          'POPESTIMATE2013',
          'POPESTIMATE2014',
          'POPESTIMATE2015']
counties_only = counties_only[['CTYNAME'] + params]
counties_only['abs_change_pop'] = (counties_only[params].max(axis=1)) - (counties_only[params].min(axis=1))
ans = counties_only[counties_only['abs_change_pop'] == max(counties_only['abs_change_pop'])]
ans.iloc[0]['CTYNAME']
```
